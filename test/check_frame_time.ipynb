```python
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
plt.style.use('presentation')
```

```python
fr1 = scipy.io.loadmat('fr_win.mat')['fr_win'].ravel()
fr2 = scipy.io.loadmat('fr_linux_nogpu.mat')['fr_linux_nogpu'].ravel()
fr3 = scipy.io.loadmat('fr_linux.mat')['fr_linux'].ravel()
```

```python
fig = plt.figure(figsize=(9, 14))
ax = plt.subplot(411)
ax.plot(fr1, ['win+gpu']*fr1.shape[0], '|', markersize=30);
ax.plot(fr2, ['lin']*fr2.shape[0], '|', markersize=30);
ax.plot(fr3, ['lin+gpu']*fr3.shape[0], '|', markersize=30);
ax.set_xlabel('Time (s)')
ax = plt.subplot(412)
ax.hist(fr1, np.linspace(0.0166, 0.0168, 1001));
ax.hist(fr2, np.linspace(0.0166, 0.0168, 1001));
ax.hist(fr3, np.linspace(0.0166, 0.0168, 1001));
ax.set_xlim(0.01665, 0.01673)
ax.set_xlabel('Time (s)')
ax.set_ylabel('# of frame')
ax = plt.subplot(413)
ax.scatter(fr1.shape[0]*['windows+Nvidia_gpu'], fr1)
ax.plot(['windows+Nvidia_gpu'], np.mean(fr1), '_', markersize=30)
ax.scatter(fr2.shape[0]*['linux+nogpu'], fr2)
ax.plot(['linux+nogpu'], np.mean(fr2), '_', markersize=30)
ax.scatter(fr3.shape[0]*['linux+AMD_gpu'], fr3)
ax.plot(['linux+AMD_gpu'], np.mean(fr3), '_', markersize=30)
ax.set_ylabel('Time (s)')
ax.set_ylim(0.01633, 0.01683)
ax = plt.subplot(414)
ax.scatter(fr1.shape[0]*['windows+Nvidia_gpu'], 1e3*(fr1 - np.mean(fr1)))
ax.plot(['windows+Nvidia_gpu'], 0, '_', markersize=30)
ax.scatter(fr2.shape[0]*['linux+nogpu'], 1e3*(fr2 - np.mean(fr2)))
ax.plot(['linux+nogpu'], 0, '_', markersize=30)
ax.scatter(fr3.shape[0]*['linux+AMD_gpu'], 1e3*(fr3 - np.mean(fr3)))
ax.plot(['linux+AMD_gpu'], 0, '_', markersize=30)
ax.set_ylabel('Time (ms)')
ax.set_ylim(-0.02, 0.02)
plt.tight_layout()
```
